```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load experiment log
df = pd.read_csv("../data/experiment_log.csv")

df.head()
```

```python
plt.figure(figsize=(6,4))
df['type'].value_counts().plot(kind='bar')
plt.title("Question Type Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

## 1. Mean similarity by question type (boxplot)

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='type', y='mean_similarity')
plt.title("Mean Similarity by Question Type")
plt.xticks(rotation=45)
plt.show()
```

## 2. Generation length by type

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='type', y='generation_length')
plt.title("Generation Length by Question Type")
plt.xticks(rotation=45)
plt.show()
```

## 3. Retrieval â†’ Generation scatter

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='mean_similarity', y='generation_length', hue='type')
plt.title("Retrieval vs Generation Effort")
plt.show()
```

## 4. Variance similarity by type

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='type', y='variance_similarity')
plt.title("Retrieval Variance by Question Type")
plt.xticks(rotation=45)
plt.show()
```

## 5. Context Attention by type

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='type', y='context_attention')
plt.title("Context Attention by Question Type")
plt.xticks(rotation=45)
plt.show()
```
